```python
#1.
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
cols = []
x=0
colx='Life expectancy '
# df=df.dropna()
for coly in df.columns:
    if colx == coly:
        continue
    if coly == 'Status' or coly=='Country':
        continue
    x = df[colx].corr(df[coly])
    cols.append((x, colx, coly))
    
            
sorted(cols, key=lambda x: x[0])
```
